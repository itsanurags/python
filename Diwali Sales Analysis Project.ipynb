# ***Exploratory Data Analysis***

```python
#calculating some statistical data like percentile, mean and std of the numerical values of the Series or DataFrame
df.describe()
```

```python
#Using describe for specific columns
df[['Age','Orders','Amount']].describe()
```

```python
#Find the total number of Male and Females #Plotting bar graph for gender column

ax= sns.countplot(x="Gender",data=df)
```

```python
ax= sns.countplot(x="Gender",data=df)

#Displaying labels for bars

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total sum of sales based on gender

sales_gen= df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen
```

```python
sns.barplot(x="Gender", y="Amount",data=sales_gen)
```

```python
sns.countplot(x="Age Group",data=df)
```

```python
sns.countplot(x="Age Group",data=df, hue="Gender")
```

```python
amt_age= df.groupby(["Age Group"], as_index=False)['Amount'].sum().sort_values(by='Age Group',ascending=True)
```

```python
amt_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=amt_age)

```

```python
top_state=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
top_state
```

```python
sns.barplot(x='State',y='Orders',data=top_state)
sns.set(rc={'figure.figsize':(15,5)})
```

```python
top_state_amt=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by='Amount',ascending=False).head(10)
```

```python
top_state_amt
```

```python
sns.barplot(x='State',y='Amount',data=top_state_amt)
sns.set(rc={'figure.figsize':(15,5)})
```
